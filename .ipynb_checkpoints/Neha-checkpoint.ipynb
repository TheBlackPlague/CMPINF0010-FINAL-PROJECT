```python
import pandas as pd
import matplotlib.pyplot as plt

read_data0 = pd.read_csv('dataset0.csv')
read_data1 = pd.read_csv('dataset1.csv')

read_data0 = read_data0[["name", "neighborhood"]]
read_data1 = read_data1[["hood", "acres"]]

read_data1.columns = ['neighborhood', 'space']

read_data0 = read_data0.groupby('neighborhood').agg({
    'name': 'nunique'
})
read_data0.columns = ['playground_count']

read_data1 = read_data1.groupby('neighborhood').agg({
    'space': 'first'
})

read_data0 = read_data0.reset_index()
read_data1 = read_data1.reset_index()

data = pd.merge(read_data0, read_data1, on='neighborhood', how='inner')
data = data.dropna()
data = data.set_index('neighborhood')
data = data.drop('Perry North', axis=0) # Faulty Data Point
data = data.reset_index()

scale_factor = 100 * (100.0 / 3);
metric = (data['playground_count'] / data['space']) * scale_factor

merged_data = pd.DataFrame({'neighborhood': data['neighborhood'], 'metric': metric})
merged_data = merged_data.sort_values(by='metric', ascending=False)
```

# Which neighborhood is the best in Pittsburgh?

## Data Sets & Metrics Used:

Analyzing the playgrounds meant that we wanted to see how many playgrounds were in a neighborhood. We wanted to be precise and be as accurate as possible given there can be so many outside factors here as well. We figured the best way was to also consider the area occupied by the neighborhood. Our measurement unit for area is acres.

Metrics:
- Number of Playgrounds in a Neighborhood
- Amount of area occupied by the Neighborhood

The specific data sets we are using here are:
- [Pittsburgh Playgrounds](https://data.wprdc.org/dataset/playgrounds)

```python
plt.figure(figsize=(15, 6))
plt.bar(data['neighborhood'], data['playground_count'])
plt.xticks(rotation=90)
plt.xlabel('Neighborhood')
plt.ylabel('Number of Playgrounds')
plt.title ('Number of Playgrounds in Neighborhoods')
plt.show()
```

We chose this dataset specifically because of its quantativeness. It gives us a clear and precise picture of the number of playgrounds in each Pittsburgh Neighborhood.

We used a bar graph to plot this metric. We believed it would be the best visualization for the metric given its quantatively infinite vertical axis, yet finite horizontal axis.

- [Pittsburgh Neighborhoods](https://data.wprdc.org/dataset/neighborhoods2)

```python
plt.figure(figsize=(15, 6))
plt.bar(data['neighborhood'], data['space'])
plt.xticks(rotation=90)
plt.xlabel('Neighborhood')
plt.ylabel('Area (acres)')
plt.title ('Area Occupied by Neighborhoods')
plt.show()
```

Likewise, we also chose this dataset because it gave us a clear picture quantatively. It accurately gives us the area (in acres) occupied by each Pittsburgh Neighborhood. Furthmore, like the previous metric, a bar graph has been used to visually represent it. 

The horizontal line has been sorted in alphabetical order for both graphs to allow easy comparison.

Looking at both of these graphs, it can be inferred that *Squirrel Hill South* has the most playgrounds in Pittsburgh, but also occupies the most area too. By both our metrics, **evaluated separately**, our judgement is very clear that *Squirrel Hill South* is the best neighborhood.

However, looking at the metrics in issolation isn't always the best idea.

"*A little knowledge is a dangerous thing.*"

#### We needed a metric that combined both our metrics. 

A neighborhood that is good in our eyes is one where the occupants don't have to walk much to get to a nearby park. As such, our proposal for this **ultimate metric** is a ratio:

 **ULTIMATE METRIC = NUMBER OF PLAYGROUNDS / AREA OCCUPIED** 

```python
plt.figure(figsize=(15, 6))
plt.bar(merged_data['neighborhood'], merged_data['metric'])
plt.xticks(rotation=90)
plt.xlabel('Neighborhood')
plt.ylabel('Quality')
plt.title ('Quality of Neighborhoods')
plt.show()
```
